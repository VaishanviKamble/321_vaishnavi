# 5. Statistical analysis (Average, Max, Min, Count, Sum, Percentage)

```python
package=studentdata[11:][0]
total_student=len(studentdata[11:][0])

# Converting String value to float
Num_package=[float(i) for i in package]
print('Average Package= ',sum(Num_package)/total_student)
```

```python
print('Minimum Package= ',min(Num_package))
```

```python
print('Maximum Package= ',max(Num_package))
```

```python
print("Math Marks=",studentdata[4])
print("Physics Marks=",studentdata[5])
print("Chemistry Marks=",studentdata[6])
# Converting String value to int
Math_Marks=[int(i) for i in studentdata[4]]
Physics_Marks=[int(i) for i in studentdata[5]]
Chemistry_Marks=[int(i) for i in studentdata[6]]

#Total marks=
Totalmarks=[]
for i in range(len(studentdata[4])):
    Totalmarks.append(Math_Marks[i]+Physics_Marks[i]+Chemistry_Marks[i])
print("Total Marks=",Totalmarks)
```

```python
percentage=[round(marks/3,2) for marks in Totalmarks]
print("Percentage=",percentage)
```
